```python
from sage.rings.continued_fraction import convergents
R.<X,Y> = QQ['X','Y'];
p = 617;
f(i)= (Y-2*X*cos(2*pi*i/p));
fphat(X,Y) = prod(f(i) for i in (1..(p-1)/2));

potentialsol = []; #collects pairs (p_k,q_k) that satisfy Fhat(p_k,q_k) = p
for k in (1..(p-1)/2):
    print("checking",k)
    convs = convergents(2*cos(2*pi*k/p)); ##should give us all the convergents 
    for conv in convs: 
        (pk,qk) = (conv.numerator(), conv.denominator());
        if qk < e^8: 
            #potentialsol.append((pk,qk));
            if ceil(abs(fphat(qk,pk).n())) ==  p or floor(abs(fphat(qk,pk).n())) ==  p:
                  potentialsol.append((qk,pk));
        else:
            print("done checking", k)
            break

print(potentialsol)
```
